```python
import sys
from pathlib import Path
from collections import Counter

ROOT = Path("C:/Users/Admin/Desktop/3차프로젝트/fa06-fin-aicc")
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from app.services.voice2 import stt2

engine = stt2.ReturnZeroSTTEngine(api_key="eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJkYSI6dHJ1ZSwiZXhwIjoxNzYzNzI1Mjg1LCJmdWUiOmZhbHNlLCJoYmkiOmZhbHNlLCJpYXQiOjE3NjM3MDM2ODUsImp0aSI6Ik9ERkYyV19DYU9tM1Q3VEdfODdDIiwicGxhbiI6ImJhc2ljIiwic2NvcGUiOiJzcGVlY2giLCJzdWIiOiJBcEtoUHNVTWxwcHBaXzZTMV9VcyIsInVjIjpmYWxzZSwidiI6MX0.0Adn4kYJnrUvLA7FdaoyqBzEVh-5U6CS2NrWFrQiL00")
svc = stt2.VitoSpeechToTextService(engine)

result = svc.transcribe_file("fixtures/sample2.mp3", language="ko", diarize=True, speaker_count=2)

# 화자 ID를 상위 2명으로 재매핑 + None을 두 화자 중 하나로 귀속
counts = Counter(s.speaker for s in result.segments if s.speaker is not None)
keep = [spk for spk, _ in counts.most_common(2)]
# 두 명보다 적게 잡혔으면 일단 하나로 둡니다
if len(keep) == 0:
    keep = [None]
elif len(keep) == 1:
    keep = [keep[0], keep[0]]

primary, secondary = keep[0], keep[1]

last = None
for s in result.segments:
    if s.speaker is None:
        # 직전 화자가 있으면 그대로, 없으면 최빈(primary)
        s.speaker = last if last is not None else primary
    else:
        # primary는 그대로, 나머지는 모두 secondary로 묶기
        s.speaker = primary if s.speaker == primary else secondary
    last = s.speaker

for s in result.segments:
    print(s.speaker, s.text)
```
